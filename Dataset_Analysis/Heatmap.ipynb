# Generation of heatmaps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load CSV file
file_path = 'DS3_NIH_Clean.csv'
df = pd.read_csv(file_path)

# Show the first few rows
df.head()
```

```python
# Drop non numeric columns
features_df = df.drop(columns=["SMILES", "Toxicity_Phenotype"])

# Calculate correlation matrix only for numeric columns
correlation_matrix = features_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="viridis",   # Try 'viridis', 'mako', 'YlGnBu', etc.
    vmin=-1, vmax=1,
    square=True
)
plt.title("Feature Correlation Heatmap (Viridis)")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="mako",   # Try 'viridis', 'mako', 'YlGnBu', etc.
    vmin=-1, vmax=1,
    square=True
)
plt.title("Feature Correlation Heatmap (Viridis)")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",  # Diverging palette
    center=0,
    vmin=-1, vmax=1,
    square=True
)
plt.title("Feature Correlation Heatmap (Coolwarm)")
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap

custom_cmap = LinearSegmentedColormap.from_list(
    "RedGoldGreen", ["crimson", "gold", "lime"]
)

plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap=custom_cmap,
    vmin=-1, vmax=1,
    square=True
)
plt.title("Feature Correlation Heatmap (Custom Redâ€“Goldâ€“Green)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

# Define colors: red â†’ lilac (a shade of violet) â†’ blue
custom_colors = ["red", "violet", "blue"]

# Create colormap
cmap = mcolors.LinearSegmentedColormap.from_list("red_lilac_blue", custom_colors, N=256)

# Plot correlation heatmap with custom colormap
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap=cmap,
    vmin=-1, vmax=1,
    square=True,
    center=0
)
plt.title("Feature Correlation Heatmap (Redâ€“Lilacâ€“Blue)")
plt.show()
```
