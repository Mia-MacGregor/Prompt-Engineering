# Dataset prep. Adding column with toxicity phenotypes and removing all NAN rows in preparation for running through models

```python
import pandas as pd

# === File Paths ===
original_csv = r'C:\PhD\Project\NIH3T3.csv'
features_csv = r'C:\PhD\Prompt_Engineering\Outputs2\DS3_Calc.csv'
output_csv = r'C:\PhD\Prompt_Engineering\Outputs2\DS3_Calc_Tox.csv'
final_output_csv = r'C:\PhD\Prompt_Engineering\Outputs2\DS3_Calc_Tox_clean.csv'

# === Load Both Datasets ===
df_original = pd.read_csv(original_csv)
df_features = pd.read_csv(features_csv)

# === Filter out rows with any "INVALID" in feature columns
feature_cols = [col for col in df_features.columns if col != 'SMILES']
df_valid = df_features[~df_features[feature_cols].isin(["INVALID", "INVALID_SMILES"]).any(axis=1)]

# === Merge on SMILES Column ===
merged_df = pd.merge(
    df_valid,
    df_original[['PUBCHEM_EXT_DATASOURCE_SMILES', 'PUBCHEM_ACTIVITY_OUTCOME']],
    how='left',
    left_on='SMILES',
    right_on='PUBCHEM_EXT_DATASOURCE_SMILES'
)

# === Clean up merged DataFrame
merged_df = merged_df.drop(columns=['PUBCHEM_EXT_DATASOURCE_SMILES'])
merged_df = merged_df.rename(columns={'PUBCHEM_ACTIVITY_OUTCOME': 'Toxicity'})

# === Save Clean Final Output
merged_df.to_csv(output_csv, index=False)
print("✅ Cleaned and merged dataset saved to:", output_csv)

# === Load and Drop Rows with Any NaNs
df = pd.read_csv(output_csv)
df_cleaned = df.dropna()
print("✅ Rows with NaN values have been removed.")

# === Save Final Cleaned Dataset
df_cleaned.to_csv(final_output_csv, index=False)
print("✅ Final cleaned dataset saved to:", final_output_csv)
```
